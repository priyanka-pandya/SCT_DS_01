```python
import pandas as pd 

# Load dataset, skip metadata rows, fix encoding issue
population_df = pd.read_csv("population.csv", skiprows=4, encoding='latin1')

# Drop fully empty rows
population_df.dropna(how='all', inplace=True)

# Fill other missing values with 0
population_df.fillna(0, inplace=True)

# Rename columns for clarity
population_df.rename(columns={
    'Country Name': 'country',
    'Country Code': 'code'
}, inplace=True)

# Remove rows where country name is missing
population_df.dropna(subset=['country'], inplace=True)

# Remove duplicates
population_df = population_df.drop_duplicates()

# Save cleaned dataset
population_df.to_csv("cleaned_data.csv", index=False)

print("Cleaned data saved successfully")
population_df.head()
```

## Histogram of Population in 2024
This visualization shows how population is distributed across different countries for the year 2024.


```python
import matplotlib.pyplot as plt
import pandas as pd 


# Load cleaned data
df = pd.read_csv("cleaned_data.csv")

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df["2024"], bins=20, edgecolor='black')
plt.title("Population Distribution in 2024")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Insights
- Most countries have populations below 100 million.
- Only a few countries exceed 500 million in population.
- The distribution is right-skewed, meaning fewer countries have very large populations.
